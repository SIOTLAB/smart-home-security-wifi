# convert time to seconds/millisecond

```python
def keep_digits(x,i):
    x = float(x[:x.find('.')+i])
    return x
```

# power per 10ms

```python
def get_attack_power_10ms(device,port_state,scenario,attack_th,attack_type,reading_numb):
    
    # Read attack file
    #file_loc1    = '/home/ydalal/media/bhagyashri/%s/power/%s/%s%s%sp%s.txt'%(device,scenario,device,attack_type,attack_th,reading_numb)
    file_loc1    = '/home/ydalal/media/bhagyashri/%s/%s/power/%s/%s%s%sp%s.txt'%(device,port_state,scenario,device,attack_type,attack_th,reading_numb)
    df1           = pd.read_csv(file_loc1,sep='\t',names=['index','timeSec','nanosecond','dontcare','voltageVolt','milliampCurrent'], low_memory=False)
    df1['power']  = df1['voltageVolt'] * df1['milliampCurrent'] 
    
    # Divide nanosecond by 1000,000,000
    df1['nanosecond'] = df1['nanosecond']/1000000000
    # # Get absolute time (0,1,2,.....,179,180)    
    df1['timeSecAbs'] =  df1['timeSec'] - df1.timeSec.min()

    # # Add seconds to nanosecond
    df1['nanosecond']   = df1['timeSecAbs'] + df1['nanosecond']

    # # # convert datatype
    df1['nanosecond'] = df1['nanosecond'].astype(str)
    df1['nanosecond']=df1['nanosecond'].apply(lambda x: keep_digits(x,3))

    # # # one row is mean of 10 ms
    df_ms_attack_power = df1[['nanosecond','power']].groupby(by=['nanosecond']).mean().reset_index()
    df_ms_attack_power.columns = ['nanosecond','power_attack']
    df_ms_attack_power.nanosecond = df_ms_attack_power.nanosecond.astype('str')
    return df_ms_attack_power
```

# attack and response per 10ms

```python
def get_attack_response_rate_10ms(device,port_state,scenario,attack_th,attack_type,reading_numb):
    out_name  = device+attack_type+str(attack_th)+'out'+str(reading_numb)
    file_path = '/home/ydalal/media/bhagyashri/%s/%s/out/%s/csv1/%s.csv'%(device,port_state,scenario,out_name)

    df        = pd.read_csv(file_path,sep='\t',names=['packets','time','ipSrc','ipDst'], low_memory=False)
    df.drop(df.index[0], inplace=True)
    df['time'] = df['time'].astype(str)
    df['time'] = df['time'].apply(lambda x: keep_digits(x,3))
    df = df.dropna()

    if attack_type == 'udp':
        
        df['ipDst'] = df['ipDst'].astype(str)
        df['ipDst_list'] = df['ipDst'].apply(lambda x: x.split(","))
        df['ipDst_count'] = df.apply(lambda x: len(x['ipDst_list']), axis=1)
        df['ipDst'] = df.apply(lambda x: x['ipDst'] if x['ipDst_count'] == 1 else x['ipDst_list'][0], axis=1)

        del df['ipDst_list']
        del df['ipDst_count']
        
        df['ipSrc'] = df['ipSrc'].astype(str)
        df['ipSrc_list'] = df['ipSrc'].apply(lambda x: x.split(","))
        df['ipSrc_count'] = df.apply(lambda x: len(x['ipSrc_list']), axis=1)
        df['ipSrc'] = df.apply(lambda x: x['ipSrc'] if x['ipSrc_count'] == 1 else x['ipSrc_list'][0], axis=1)

        del df['ipSrc_list']
        del df['ipSrc_count']
        temp1 = df.query('ipDst.str.contains("192.168.1.245") and packets.str.contains("Frame is not being retransmitted")')
        temp2 = df.query('ipDst.str.contains("192.168.1.245") and packets.str.contains("Frame is being retransmitted")')
        temp3 = df.query('ipDst.str.contains("192.168.1.245")' )
        temp4 = df.query('ipSrc.str.contains("192.168.1.245")')
    else:
        # destination is alexa and packet is unique
        temp1 = df.query('ipDst.str.contains("192.168.1.245") and packets.str.contains("Frame is not being retransmitted")')
        # destination is alexa and packet is restransmitted
        temp2 = df.query('ipDst.str.contains("192.168.1.245") and packets.str.contains("Frame is being retransmitted")')
        # destination is alexa and packet is unique or restransmitted
        temp3 = df.query('ipDst.str.contains("192.168.1.245")' )
        # source is alexa i,e response
        temp4 = df.query('ipSrc.str.contains("192.168.1.245")')
    
    df1 = temp1.groupby(by=['time','ipDst'])['ipDst'].count().\
        to_frame('unique_attack_rate').reset_index()

    df2 = temp2.groupby(by=['time','ipDst'])['ipDst'].count().\
        to_frame('retran_attack_rate').reset_index()

    df3 = temp3.groupby(by=['time','ipDst'])['ipDst'].count().\
        to_frame('total_attack_rate').reset_index()

    df4 = temp4.groupby(by=['time','ipSrc'])['ipSrc'].count().\
        to_frame('total_response_rate').reset_index()
    
    df1.time = df1.time.astype('str')
    df2.time = df2.time.astype('str')
    df3.time = df3.time.astype('str')
    df4.time = df4.time.astype('str')

    df_merge_0 = df1.merge(df2,on=['time','ipDst'],how='outer').fillna(0)
    df_merge_1 = df_merge_0.merge(df3,on=['time','ipDst'],how='outer').fillna(0)
    df_merge_2 = df_merge_1.merge(df4,on=['time'],how='outer').fillna(0)
#     print ("temp4 is")
#     print temp4
    return df_merge_2
```

#  attack rate and response and power per 10ms

```python
def get_attack_response_rate_power_10ms(device,port_state,scenario,attack_th,attack_type,reading_numb):
    df_ms_power  = get_attack_power_10ms(device,port_state,scenario,attack_th,attack_type,reading_numb)
#     print df_ms_power.head()
    df_ms_rate   = get_attack_response_rate_10ms(device,port_state,scenario,attack_th,attack_type,reading_numb)
#     print df_ms_rate.head()
    df_ms_merged = pd.merge(df_ms_power,df_ms_rate,left_on=['nanosecond'],
                                                            right_on=['time'],
                                                            how='inner')
    return df_ms_merged
```

# Attack rate and power per 500ms

```python
def get_attack_response_rate_power_500ms(df_ms_merged):
              
    # Since out of 1 second interval the shift is for x ms, the 1sec interval rate and power will have 
    # small value becasue not complete.
    idle_power = 556.827584196
    frames = []
    rows   = []

    # power per 500ms
    df_500ms_power = df_ms_merged[['nanosecond', 'power_attack']].copy(deep=True)
    df_500ms_power['nanosecond'] = df_500ms_power['nanosecond'].astype(str)
    df_500ms_power['nanosecond'] =  df_500ms_power['nanosecond'].apply(lambda x: keep_digits(x,1))

    for i in df_500ms_power.nanosecond.unique():
        df = df_500ms_power[df_500ms_power['nanosecond']== i]
        center_index = df.shape[0]/2
        if center_index == 0:
            center_index=1

        power_mean_per_500ms = df[:center_index]['power_attack'].mean()
        frames.append([i,power_mean_per_500ms])

        power_mean_per_500ms = df[center_index:]['power_attack'].mean()
        frames.append([i+0.500,power_mean_per_500ms])


    df_500ms_power = pd.DataFrame(frames,columns=['nanosecond','power_attack'])
    df_500ms_power = df_500ms_power.dropna()
    df_500ms_power['power_perc_chng'] = ((df_500ms_power['power_attack'] - idle_power)/idle_power)*100

    
    
    # attack_rate per 500ms
    df_500ms_attack_rate = df_ms_merged[['time','unique_attack_rate','retran_attack_rate','total_attack_rate','total_response_rate']].copy(deep=True)
    df_500ms_attack_rate['time'] = df_500ms_attack_rate['time'].astype(str)
    df_500ms_attack_rate['time'] =  df_500ms_attack_rate['time'].apply(lambda x: keep_digits(x,1))
    
    for i in df_500ms_attack_rate.time.unique():
        df = df_500ms_attack_rate[df_500ms_attack_rate['time']==i]
        center_index = int (df.shape[0]/2)
        if center_index == 0:
            center_index=1
        
        total_rate_count_per_500ms = df[:center_index]['total_attack_rate'].sum()
        unique_rate_count_per_500ms = df[:center_index]['unique_attack_rate'].sum()
        response_rate_count_per_500ms = df[:center_index]['total_response_rate'].sum()
        rows.append([i,total_rate_count_per_500ms,unique_rate_count_per_500ms,response_rate_count_per_500ms])
        

        total_rate_count_per_500ms = df[center_index:]['total_attack_rate'].sum()
        unique_rate_count_per_500ms = df[center_index:]['unique_attack_rate'].sum()
        response_rate_count_per_500ms = df[center_index:]['total_response_rate'].sum()

        rows.append([i+0.500,total_rate_count_per_500ms,unique_rate_count_per_500ms,response_rate_count_per_500ms])
        
        df_500ms_rate = pd.DataFrame(rows,columns=['time','total_rate_count','unique_rate_count','response'])
        df_500ms_rate['uniq%'] =  (df_500ms_rate['total_rate_count']-df_500ms_rate['unique_rate_count'])/df_500ms_rate['total_rate_count']
        df_500ms_rate['uniq%'] = df_500ms_rate['uniq%']*100
    
    df_500ms_attack_response_rate = pd.DataFrame(rows,columns=['time','total_attack_rate','unique_attack_rate','total_response_rate'])
    df_500ms_attack_response_power_merged = pd.merge(df_500ms_power,df_500ms_attack_response_rate,left_on=['nanosecond'],right_on=['time'],how='inner')
    return df_500ms_attack_response_power_merged
```

```python
rows = []

df_ms_merged = df_ms_merged_3.copy(deep=True)


df_ms_merged = df_ms_merged.shift(200)
df_ms_merged = df_ms_merged.dropna()
df_ms_merged.time = df_ms_merged.time.astype('float')
df_ms_merged['time'] =  df_ms_merged['time'] - df_ms_merged.time.min()

df_500ms_rate = df_ms_merged.copy(deep=True)
df_500ms_rate['time'] = df_500ms_rate['time'].astype(str)
df_500ms_rate['time'] =  df_500ms_rate['time'].apply(lambda x: keep_digits(x,1))


for i in df_500ms_rate.time.unique():
    df = df_500ms_rate[df_500ms_rate['time']==i]
    center_index = int (df.shape[0]/2)
    if center_index == 0:
        center_index=1
    total_rate_count_per_500ms = df[:center_index]['total_attack_rate'].sum()
    unique_rate_count_per_500ms = df[:center_index]['unique_attack_rate'].sum()
    response_rate_count_per_500ms = df[:center_index]['total_response_rate'].sum()
    
    rows.append([i,total_rate_count_per_500ms,unique_rate_count_per_500ms,response_rate_count_per_500ms])

    total_rate_count_per_500ms = df[center_index:]['total_attack_rate'].sum()
    unique_rate_count_per_500ms = df[center_index:]['unique_attack_rate'].sum()
    response_rate_count_per_500ms = df[center_index:]['total_response_rate'].sum()

    rows.append([i+0.500,total_rate_count_per_500ms,unique_rate_count_per_500ms,response_rate_count_per_500ms])

df_500ms_rate = pd.DataFrame(rows,columns=['time','total_rate_count','unique_rate_count','response'])
df_500ms_rate['uniq%'] =  (df_500ms_rate['total_rate_count']-df_500ms_rate['unique_rate_count'])/df_500ms_rate['total_rate_count']
df_500ms_rate['uniq%'] = df_500ms_rate['uniq%']*100

# df_500ms_merged.drop(df_500ms_merged.tail(1).index,inplace=True)
# corr = stats.pearsonr(df_500ms_merged['attack_rate'],df_500ms_merged['power_attack'])[0]
# print ("corr is: " +str(corr))
# # df_ms_merged
# # print (df_500ms_power.tail(20))
# # print (df_500ms_rate.tail(20))
# #print (df_500ms_merged.tail(20))
# print (df_500ms_merged.head(176))
df_500ms_rate
```
